```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_metrics(sample: np.ndarray) -> None:
    print(f'Mathematical expectation: {np.median(sample)}')
    print(f'Dispersion: {np.var(sample).round(3)}')
    print(f'Sigma: {np.std(sample).round(3)}')
```

## Constants

```python
default_distribution_size = int(10e3)
```

## Building uniform distribution

```python
small_distribution_size = 100
uniform_distribution = np.random.uniform(-50, 50, small_distribution_size).astype(int)

plt.hist(uniform_distribution, bins=20, color='r')
plt.show()
calculate_metrics(uniform_distribution)

uniform_distribution = np.random.uniform(-50, 50, default_distribution_size).astype(int)

plt.hist(uniform_distribution, bins=20, color='r')
plt.show()
calculate_metrics(uniform_distribution)
```

## Correcting the distribution

```python
coefficient = 0.0000000000001
correction_array = np.arange(uniform_distribution.shape[0])
correction_array = (correction_array ** 3) * coefficient

corrected_distribution = uniform_distribution + correction_array

plt.plot(corrected_distribution)
plt.plot(correction_array, color='red')
plt.show()

calculate_metrics(corrected_distribution)
```

## Let's compare

```python
plt.hist(uniform_distribution, bins=20, alpha=0.5)
plt.hist(corrected_distribution, bins=20, alpha=0.5)

plt.show()
```

## Take in consideration the dinamic of change

```python
corrected_back_distribution = corrected_distribution - correction_array

plt.hist(uniform_distribution, bins=20, alpha=0.5)
plt.hist(corrected_distribution, bins=20, alpha=0.5)
plt.hist(corrected_back_distribution, bins=20, alpha=0.5)

plt.show()
```

## Normal Distribution

```python
mean, sigma = 0, 10
normal_distribution = np.random.normal(mean, sigma, size=default_distribution_size).astype(int)

plt.hist(normal_distribution, bins=20, color='r')
plt.show()
calculate_metrics(normal_distribution)
```

## Correcting the distribution

```python
coefficient = 0.004
correction_norm_array = np.arange(normal_distribution.shape[0])
correction_norm_array = correction_norm_array * coefficient

corrected_norm_distribution = normal_distribution + correction_norm_array

plt.plot(corrected_norm_distribution)
plt.plot(correction_norm_array)
plt.show()

calculate_metrics(corrected_norm_distribution)
```

## Let's compare

```python
plt.hist(normal_distribution, bins=20, alpha=0.5)
plt.hist(corrected_norm_distribution, bins=20, alpha=0.5)

plt.show()
```

## Take in consideration the dinamic of change

```python
corrected_back_norm_distribution = corrected_norm_distribution - correction_norm_array

plt.hist(normal_distribution, bins=20, alpha=0.5)
plt.hist(corrected_norm_distribution, bins=20, alpha=0.5)
plt.hist(corrected_back_norm_distribution, bins=20, alpha=0.5)

plt.show()
```
